```python
from PIL import Image, ImageOps
import numpy as np
import matplotlib.pyplot as plt
```

```python

# 画像の読み込み & グレースケールに変換
img = Image.open('imgs/girl.jpg').convert('L')

# 上段の画像の表示設定
plt.subplot(2, 1, 1) # 引数はそれぞれ、全体の行数、全体の列数、設定対象のIndex
a = np.array(img)
plt.imshow(a)
plt.gray()
plt.axis('off')
plt.show()
```

```python
# 256階級のヒストグラムの描画
plt.subplot(2, 2, 1)
plt.hist(a.flatten(), bins=np.arange(256 )) #階級の幅を256としてヒストグラムを出す。
# 図表の表示
plt.show()
```

```python
#8階級のヒストグラム
plt.subplot(2,2,1)
plt.hist(a.flatten(), bins=np.arange(8)) #階級の幅を８にしてヒストグラムを出す。
plt.show()
```

```python
img = Image.open('imgs/overlight.jpg').convert('L')
# 上段の画像の表示設定
plt.subplot(2, 1, 1) # 引数はそれぞれ、全体の行数、全体の列数、設定対象のIndex
a = np.array(img)
plt.imshow(a)
plt.gray()
plt.axis('off')
plt.show()
```

```python
plt.subplot(2, 2, 1) # 引数はそれぞれ、全体の行数、全体の列数、設定対象のIndex
plt.hist(a.flatten(), bins=np.arange(256)) #階級の幅を８にしてヒストグラムを出す。
plt.show()
```

```python
plt.subplot(2, 2, 1) # 引数はそれぞれ、全体の行数、全体の列数、設定対象のIndex
img = ImageOps.equalize(img);
a = np.array(img)
plt.imshow(img)
plt.gray()
plt.axis('off')
plt.subplot(2, 2, 3) # 引数はそれぞれ、全体の行数、全体の列数、設定対象のIndex
plt.hist(a.flatten(), bins=np.arange(256)) 
plt.show()
```

演習課題２

```python
images = ['imgs/shimauma.jpg', 'imgs/usagi.jpg', 'imgs/water.jpg', 'imgs/ct.jpg', 'imgs/house.jpg']
for index in range(5):
  img = Image.open(images[index]).convert('L')
  a = np.array(img)
  plt.subplot(2, 5, index+1) 
  plt.imshow(img)
  plt.gray()
  plt.axis('off')
  plt.subplot(2, 5, index+6) # 引数はそれぞれ、全体の行数、全体の列数、設定対象のIndex
  plt.hist(a.flatten(), bins=np.arange(256)) 

plt.show()
  
```
